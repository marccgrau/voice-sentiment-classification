```python
import os
import numpy as np
import librosa


def is_audio_empty(audio_path):
    y, sr = librosa.load(audio_path, sr=None)
    return np.all(y == 0)


def check_empty_audios(directory):
    empty_files = []
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith(".wav"):
                file_path = os.path.join(root, file)
                if is_audio_empty(file_path):
                    empty_files.append(file_path)
    return empty_files


directories = ["raw_data/test", "raw_data/train", "raw_data/valid"]
for directory in directories:
    empty_files = check_empty_audios(directory)
    if empty_files:
        print(f"Empty audio files in {directory}:")
        for file in empty_files:
            print(file)
    else:
        print(f"No empty audio files found in {directory}.")
```

```python
for directory in directories:
    empty_files = check_empty_audios(directory)
    if empty_files:
        print(f"Empty audio files in {directory}:")
        for file in empty_files:
            print(file)
            os.remove(file)  # Delete the empty file
        print(f"Deleted {len(empty_files)} empty audio files in {directory}.")
    else:
        print(f"No empty audio files found in {directory}.")
```

```python
for directory in directories:
    total_files = 0
    for root, _, files in os.walk(directory):
        total_files += len(files)
    print(f"Total number of files in {directory}: {total_files}")
```
